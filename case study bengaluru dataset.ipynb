## outlier treatment
- ### iqr method
- iqr=q3-q1
- q3= 75 th percentile
- q1= 25th percentile


```python
#creating a function for outlier treatment

def outlier(data):

    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)

    iqr = q3-q1
    upper_bound = q3+(1.5*(iqr))
    lower_bound = q1-(1.5*(iqr))


    return data.clip(upper_bound, lower_bound)
```

### df.corr()  
 - since our dataset has numeric and categorical columns we can't do df.corr() in jupyter notebook the only way to do df.corr() and thereby to plot the heatmap is by considering only the numeric values in the dataset by using a parameter numeric_only=True.First we need to create a correlation matrix of the dataset and then pass the parameter numeric_only=True.
  - can do on google colabatory df.corr() for both numeric and categorical data


```python
corr_mat=df.corr(numeric_only=True)
corr_mat
```

```python
sns.heatmap(corr_mat,annot=True)
plt.show()
```
